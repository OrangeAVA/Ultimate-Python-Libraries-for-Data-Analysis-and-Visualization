```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# Reading a CSV file into a Pandas DataFrame
data = pd.read_csv('IPL_Matches_2008_2022.csv')
# Checking the first few rows of the DataFrame
data.head()
```

```python
data.shape
```

```python
# Check for missing values
print(data.isnull().sum())
```

```python
sns.histplot(data['Margin'])
plt.show()
```

```python
sns.boxplot(data['Margin'])
plt.show()
```

```python
skewness = data['Margin'].skew()
print("Skewness:", skewness)
```

**Interpretation:**

- Skewness > 0: Right-skewed distribution
- Skewness < 0: Left-skewed distribution
- Skewness close to 0: Approximately symmetric distribution

```python

# Fill missing values with the median (Imputation)

data['Margin'].fillna(data['Margin'].median(), inplace=True)  # Impute with median
data.head()
```

```python
# Drop missing values
data_drop = data.dropna(subset= ['method'])
data_drop.head()
```

```python
#Outlier detection using box plot
# Plot the box plot
sns.boxplot(data['Margin'])
```

```python
#Outlier detection
#IQR Technique
# Plot the distribution of the 'Margin' column
plt.hist(data['Margin'])
plt.xlabel("Margin")
plt.ylabel("Frequency")
plt.title("Distribution of Margin Values")
plt.show()
```

```python
# Identify outliers using the Interquartile Range (IQR)
Q1 = data['Margin'].quantile(0.25)
Q3 = data['Margin'].quantile(0.75)
IQR = Q3 - Q1

# Define outliers as values outside the 1.5 IQR range
outliers = data[data['Margin'] < Q1 - 1.5 * IQR]
outliers = pd.concat([outliers,(data[data['Margin'] > Q3 + 1.5 * IQR])])
# Print the outliers
outliers
```

```python
outliers.head()
```

```python

# Calculate the z-scores
z_scores = np.abs(data['Margin'] - data['Margin'].mean()) / data['Margin'].std()
# Define outliers as data points that are more than 3 standard deviations away from the mean
outliers = z_scores[z_scores > 3]
outliers

```

```python
data.iloc[list(outliers.index)]
```

```python
import pandas as pd

# Reading a CSV file into a Pandas DataFrame
data = pd.read_csv('insurance.csv')

# Checking the first few rows of the DataFrame
data.head()
```

```python
# Applying a function to a column
data['new_charges'] = data['charges'].apply(lambda x: x * 2)
data['new_charges']
```

```python
data
```

```python
# Create a new feature called 'AgeGroup'
data['AgeGroup'] = data['age'].apply(lambda x: 'Young' if x < 30 else 'Adult' if x < 50 else 'Senior')
```

```python
data['AgeGroup']
```

```python
# One-hot encode the 'Smoker' column
data = pd.get_dummies(data, columns=['smoker'], dtype = int)
```

```python
data
```

```python
# standardize the Series
s_standardized = (data['age'] - data['age'].mean()) / data['age'].std()
print(s_standardized)
```

```python
# normalize the Series
s_normalized = (data['age'] - data['age'].min()) / (data['age'].max() - data['age'].min())
print(s_normalized)
```

```python
#Log Transform 

import numpy as np
import matplotlib.pyplot as plt

# Perform the log transformation
data_log = np.log(data['charges'])

# Define a grid of plots
fig, axs = plt.subplots(nrows=1, ncols=2)

# Create histograms
axs[0].hist(data['charges'], edgecolor='black')
axs[1].hist(data_log, edgecolor='black')

# Add a title to each histogram
axs[0].set_title('Original Data')
axs[1].set_title('Log-Transformed Data')

# Display the plots
plt.show()
```

```python
#Square Root Transform

import numpy as np
import matplotlib.pyplot as plt

# Perform the square root transformation
data_sqrt = np.sqrt(data['charges'])

# Define a grid of plots
fig, axs = plt.subplots(nrows=1, ncols=2)

# Create histograms
axs[0].hist(data['charges'], edgecolor='black')
axs[1].hist(data_sqrt, edgecolor='black')

# Add a title to each histogram
axs[0].set_title('Original Data')
axs[1].set_title('Square Root Transformed Data')

# Display the plots
plt.show()
```

```python
#Concatenation

import pandas as pd

df1 = pd.DataFrame({'Column1': ['A0', 'A1', 'A2'],
                    'Column2': ['B0', 'B1', 'B2'],
                    'Column3': ['C0', 'C1', 'C2'],
                    'Column4': ['D0', 'D1', 'D2']},
                   index=[0, 1, 2])

df2 = pd.DataFrame({'Column1': ['A3', 'A4', 'A5'],
                    'Column2': ['B3', 'B4', 'B5'],
                    'Column3': ['C3', 'C4', 'C5'],
                    'Column4': ['D3', 'D4', 'D5']},
                   index=[3, 4, 5])

result = pd.concat([df1, df2])

result
```

```python
#Merge

# Creating the first DataFrame
df1 = pd.DataFrame({
    'Employee': ['Tom', 'Ben', 'Tom','Suzy'],
    'Department': ['IT', 'Marketing', 'IT', 'R&D']
})

# Creating the second DataFrame
df2 = pd.DataFrame({
    'Employee': ['Tom', 'Suzy'],
    'Salary': [8000, 10000]
})

merged_df = pd.merge(df1, df2, on='Employee')
merged_df
```

```python
merged_df = pd.merge(df1, df2, on='Employee', how='left')
merged_df
```

```python
#Join

left = pd.DataFrame({'A': ['A0', 'A1', 'A2']}, index=['x', 'y', 'z'])
right = pd.DataFrame({'B': ['B0', 'B1', 'B2']}, index=['x', 'a', 'b'])
joined = left.join(right)
joined
```
